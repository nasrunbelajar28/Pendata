# **Mengukur Jarak Data**

Mengukuran jarak data (Similarity dan Dissimilarity) adalah proses untuk menentukan seberapa mirip atau berbeda dua data atau lebih. Jarak ini dapat digunakan dalam berbagai algoritma pembelajaran mesin, seperti clustering, klasifikasi, dan regresi. Dalam mengukur jarak data terdapat beberapa tipe atribut seperti:

- Numerik

- Binary

- Categorical

- Ordinal

## **Mengukur Jarak Data (Tipe Numerik)**

Salah satu tantangan dalam era ini adalah database yang memiliki banyak tipe data. Mengukur jarak adalah komponen utama dalam algoritma clustering berbasis jarak. Algoritma partitioning seperti K-Means, K-Medoids, Fuzzy C-Means, dan Rough Clustering sangat bergantung pada perhitungan jarak untuk melakukan pengelompokkan data.

Sebelum menjelaskan tentang beberapa macam ukuran jarak, kita mendefinisikan terlebih dahulu yaitu $v_1$ dan $v_2$ menyatakan dua vektor:
$$
v_1 = (x_1, x_2, \ldots, x_n)
$$
$$
v_2 = (y_1, y_2, \ldots, y_n)
$$di mana $x_i$ dan $y_i$ disebut atribut. Ada beberapa ukuran similaritas atau ukuran jarak, di antaranya:

##**1. Minkowski Distance**
Minkowski adalah ukuran jarak umum yang mencakup Euclidean dan Manhattan sebagai kasus khusus.

$$
d(x,y) = \left( \sum_{i=1}^{n} |x_i - y_i|^m \right)^{\frac{1}{m}}, \quad m \ge 1
$$

- $m=1$ → Manhattan  
- $m=2$ → Euclidean  


##**2. Manhattan Distance**
Menghitung jarak sebagai jumlah selisih absolut tiap atribut.

$$
d(x,y) = \sum_{i=1}^{n} |x_i - y_i|
$$

##**3. Euclidean Distance**  
Jarak garis lurus antar dua titik, paling sering digunakan.

$$
d(x,y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

Kelemahan: sensitif terhadap skala besar dan outlier.

##**4. Average Distance**  
Versi rata-rata dari Euclidean untuk mengurangi pengaruh jumlah dimensi.

$$
d(x,y) = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - y_i)^2}
$$

##**5. Weighted Euclidean Distance**  
Memberikan bobot pada setiap atribut.

$$
d(x,y) = \sqrt{\sum_{i=1}^{n} w_i (x_i - y_i)^2}
$$


##**6. Chord Distance**  
Modifikasi Euclidean berbasis normalisasi panjang vektor.

$$
d(x,y) = \sqrt{2 - 2 \frac{\sum x_i y_i}{\|x\| \|y\|}}
$$

##**7. Mahalanobis Distance**  
Memperhitungkan korelasi antar atribut (lebih akurat untuk data kompleks).

$$
d(x,y) = \sqrt{(x - y)^T S^{-1} (x - y)}
$$


##**8. Cosine Similarity**  
Mengukur kemiripan arah (sering dipakai pada teks/dokumen).

$$
\cos(x,y) = \frac{\sum x_i y_i}{\|x\| \|y\|}
$$


##**9. Pearson Correlation**  
Mengukur hubungan linear antar dua data.

$$
r = \frac{\sum (x_i - \mu_x)(y_i - \mu_y)}{\sqrt{\sum (x_i - \mu_x)^2} \sqrt{\sum (y_i - \mu_y)^2}}
$$

Kelemahan: sensitif terhadap outlier.

##**Jarak Atribut Binary**

##**Biner Simetris**
Biner simetris adalah atribut biner di mana kedua nilai (0 dan 1) sama pentingnya. Nilai 1 tidak lebih penting dari 0, begitupun sebaliknya. Rumus menghitung jarak data dengan atribut biner simetris sebagai berikut:



$$
d(x,y) = \sum_{i=1}^{n} |x_i - y_i|
$$

Dimana:
- $x$ merupakan data pertama  
- $y$ merupakan data kedua  
- $x_i$ merupakan fitur ke-$i$ dari data pertama  
- $y_i$ merupakan fitur ke-$i$ dari data kedua  

#**Mengukur Jarak Tipe Categorical (Nominal)**
Digunakan untuk data dengan atribut kategori (misalnya: warna, jenis, dll).

#**1. Overlay Metric (OM)**#  
Menghitung jarak dengan cara membandingkan setiap atribut.

$$
d(x,y) = \sum_{i=1}^{n} \delta(a_i(x), a_i(y))
$$

- $\delta = 0$ jika sama  
- $\delta = 1$ jika berbeda  

Intinya: hitung berapa banyak atribut yang berbeda.

#**2. Value Difference Metric (VDM)**  
Mengukur jarak berdasarkan perbedaan probabilitas kelas dari tiap nilai atribut.

$$
d(x,y) = \sum_{i=1}^{n} \sum_{c=1}^{C} |P(c|a_i(x)) - P(c|a_i(y))|
$$

Intinya: dua nilai dianggap mirip jika sering muncul pada kelas yang sama.


#**3. Short and Fukunaga Metric (SFM)**  
Mengukur jarak berdasarkan probabilitas keseluruhan data.

$$
d(x,y) = \sum_{c=1}^{C} |P(c|x) - P(c|y)|
$$

Probabilitas biasanya dihitung dengan Naive Bayes.

#**4. Minimum Risk Metric (MRM)**  
Mengukur jarak berdasarkan risiko kesalahan klasifikasi.

$$
d(x,y) = \sum_{c=1}^{C} P(c|x)(1 - P(c|y))
$$

Intinya: semakin kecil risiko salah klasifikasi, semakin mirip datanya.

#**Mengukur Jarak Tipe Ordinal**  

Atribut ordinal adalah data yang memiliki urutan (ranking), misalnya: kecil < sedang < besar.  
Namun, jarak antar nilai tidak diketahui pasti.


Langkah Menghitung Jarak:

#**1. Ubah nilai kategori menjadi peringkat**  
Misalnya:
kecil = 1, sedang = 2, besar = 3  

$$
r_{if} \in \{1, 2, ..., M_f\}
$$


#**2. Normalisasi peringkat ke rentang [0,1]**  
Agar semua atribut memiliki skala yang sama:

$$
z_{if} = \frac{r_{if} - 1}{M_f - 1}
$$


#**3. Hitung jarak seperti data numerik**  
Setelah menjadi $z_{if}$, gunakan jarak seperti Euclidean atau Manhattan.


Intinya:  
- Data ordinal → ubah ke ranking  
- Ranking → dinormalisasi  
- Setelah itu → dihitung seperti data numerik

#**Menghitung Jarak Tipe Campuran**
Digunakan untuk data yang memiliki berbagai tipe atribut:
numerik, nominal, biner, dan ordinal.

#**Rumus Umum:**

$$
d(i,j) = \frac{\sum_{f=1}^{p} \delta_{ij}^{(f)} \, d_{ij}^{(f)}}{\sum_{f=1}^{p} \delta_{ij}^{(f)}}
$$


#**Keterangan:**
- $p$ = jumlah atribut  
- $\delta_{ij}^{(f)} = 0$ jika data hilang / tidak valid  
- $\delta_{ij}^{(f)} = 1$ jika data valid  
- $d_{ij}^{(f)}$ = jarak pada atribut ke-$f$


#**Cara Hitung Tiap Tipe Atribut:**

#1. Numerik  
$$
d_{ij}^{(f)} = \frac{|x_{if} - x_{jf}|}{\max(x_f) - \min(x_f)}
$$
Selisih dinormalisasi (agar skala sama)



#2. Nominal / Biner  
$$
d_{ij}^{(f)} =
\begin{cases}
0, & \text{jika sama} \\
1, & \text{jika berbeda}
\end{cases}
$$

Sama = 0, beda = 1



#3. Ordinal  
Ubah ke ranking lalu normalisasi:

$$
z_{if} = \frac{r_{if} - 1}{M_f - 1}
$$

Setelah itu diperlakukan seperti numerik



Intinya:
- Semua tipe data diubah ke skala yang sama (0–1)
- Hitung jarak tiap atribut
- Ambil rata-rata → jadi jarak total

Ini sering disebut juga sebagai Gower Distance

#**Memuat Dataset**

```python
df = pd.read_csv("StudentsPerformance (1).csv")
df.head()
```

Data diatas didapatkan dari platform https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?resource=download

Dataset ini berisi:

- Numerik → math score, reading score, writing score

- Kategori → gender, race, lunch

- Ordinal (bisa dibuat) → tingkat pendidikan

#**Data Pembanding**
Kita ambil 2 baris sebagai contoh perhitungan jarak.

```python
data_1 = df.iloc[0]
data_2 = df.iloc[1]

print(data_1)
print(data_2)
```

#**1. TIPE NUMERIK**
ambil fitur numerik

```python
num_1 = data_1[['math score','reading score','writing score']].values
num_2 = data_2[['math score','reading score','writing score']].values

print(num_1)
print(num_2)
```

Penjelasan:

Hanya ambil nilai angka untuk dihitung jaraknya.

#**Euclidean Distance**

```python
euc_scipy = distance.euclidean(num_1, num_2)
euc_numpy = np.linalg.norm(num_1 - num_2)
euc_manual = sqrt(sum((x - y)**2 for x, y in zip(num_1, num_2)))

print("Euclidean:", euc_scipy, euc_numpy, euc_manual)
```

Penjelasan:

Mengukur jarak lurus antar data.

#**Manhattan Distance**

```python
man_scipy = distance.cityblock(num_1, num_2)
man_numpy = np.sum(np.abs(num_1 - num_2))
man_manual = sum(abs(x - y) for x, y in zip(num_1, num_2))

print("Manhattan:", man_scipy, man_numpy, man_manual)
```

Penjelasan:

Menjumlahkan selisih tiap fitur.

#**2. TIPE CATEGORICAL**
Ambil Data Kategori

```python
cat_1 = data_1[['gender','race/ethnicity','lunch']]
cat_2 = data_2[['gender','race/ethnicity','lunch']]

print(cat_1)
print(cat_2)
```

Penjelasan:

Kategori tidak bisa langsung dihitung → pakai metode khusus.

#**Overlay Metric**

```python
def overlay_metric(p1, p2):
    delta = [0 if a == b else 1 for a, b in zip(p1, p2)]
    total = sum(delta)
    return total, total/len(p1)

jarak, norm = overlay_metric(cat_1, cat_2)

print("OM:", jarak)
print("Normalisasi:", norm)
```

Penjelasan:

Sama = 0

Beda = 1

Dijumlahkan → jadi jarak

#**3. TIPE ORDINAL**
Ubah ke ranking

```python
ranking = {
    "some high school": 1,
    "high school": 2,
    "some college": 3,
    "associate's degree": 4,
    "bachelor's degree": 5,
    "master's degree": 6
}

ord_1 = ranking[data_1['parental level of education']]
ord_2 = ranking[data_2['parental level of education']]
```

Penjelasan:

Kategori diubah ke angka berdasarkan urutan.

**Normalisasi**

```python
M = max(ranking.values())

z1 = (ord_1 - 1)/(M - 1)
z2 = (ord_2 - 1)/(M - 1)

print(z1, z2)
```

Penjelasan:

Agar semua nilai berada di skala 0–1.

#**Hitung Jarak**

```python
jarak_ord = abs(z1 - z2)
print("Jarak Ordinal:", jarak_ord)
```

Penjelasan:

Setelah normalisasi → dihitung seperti numerik.

#**4. TIPE CAMPURAN (MIXED)**
Hitung Jarak Campuran

```python
# Range numerik untuk normalisasi
range_math = df['math score'].max() - df['math score'].min()

def jarak_campuran(p1, p2):
    # Numerik
    d_num = abs(p1['math score'] - p2['math score']) / range_math

    # Kategori
    d_cat = 0 if p1['gender'] == p2['gender'] else 1

    # Ordinal
    r1 = ranking[p1['parental level of education']]
    r2 = ranking[p2['parental level of education']]
    z1 = (r1 - 1)/(M - 1)
    z2 = (r2 - 1)/(M - 1)
    d_ord = abs(z1 - z2)

    return (d_num + d_cat + d_ord)/3

hasil = jarak_campuran(data_1, data_2)
print("Jarak Campuran:", hasil)
```

Penjelasan:

Numerik → selisih normalisasi

Kategori → sama/beda

Ordinal → ranking

Semua dirata-rata
