UAS ini bertujuan untuk menerapkan algoritma **K-Nearest Neighbor (KNN)** menggunakan KNIME untuk melakukan klasifikasi data mahasiswa. # 2. Dataset

Dataset yang digunakan merupakan file **DATA (1).csv** dengan:

- Jumlah data: **145 baris**
- Jumlah atribut: **33 kolom**

Kolom terdiri dari:
- STUDENT ID
- Pertanyaan 1–30
- COURSE ID
- GRADE
