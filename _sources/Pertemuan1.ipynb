# Pertemuan 1

## Apa itu Penambangan Data?

Penambangan Data (atau Data Mining) adalah proses teknik dan metode untuk menggali informasi serta pola tersembunyi dari sekumpulan data yang besar. Tujuannya adalah untuk mengubah data mentah menjadi wawasan berharga yang dapat mendukung pengambilan keputusan secara lebih akurat dan efektif.

## Pengertian Aplikasi Tools Data Mining

Aplikasi-aplikasi ini digunakan untuk memproses, menganalisis, dan memodelkan data :

KNIME (Konstanz Information Miner): Platform analisis data berbasis "drag-and-drop". Kamu bisa membuat alur kerja (workflow) data dari mulai pembersihan data, manipulasi, hingga pemodelan statistik tanpa harus mahir menulis kode pemrograman yang rumit.


Orange: Mirip dengan KNIME, Orange adalah perangkat lunak sumber terbuka (open-source) untuk visualisasi data dan pembelajaran mesin (machine learning). Tampilannya sangat visual, sehingga cocok bagi pemula untuk melihat pola data secara langsung.

## Pengertian Aplikasi Database (Penyimpanan Data)
Aplikasi ini digunakan untuk menyimpan dan mengelola data dalam skala besar:


PostgreSQL: Sistem manajemen database relasional (RDBMS) yang sangat canggih dan bersifat open-source. PostgreSQL dikenal karena ketangguhannya dalam menangani data yang kompleks dan besar dengan sangat stabil.
