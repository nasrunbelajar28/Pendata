# **Mengukur Jarak Data**

Mengukuran jarak data (Similarity dan Dissimilarity) adalah proses untuk menentukan seberapa mirip atau berbeda dua data atau lebih. Jarak ini dapat digunakan dalam berbagai algoritma pembelajaran mesin, seperti clustering, klasifikasi, dan regresi. Dalam mengukur jarak data terdapat beberapa tipe atribut seperti:

- Numerik

- Binary

- Categorical

- Ordinal

## **Mengukur Jarak Data (Tipe Numerik)**

Salah satu tantangan dalam era ini adalah database yang memiliki banyak tipe data. Mengukur jarak adalah komponen utama dalam algoritma clustering berbasis jarak. Algoritma partitioning seperti K-Means, K-Medoids, Fuzzy C-Means, dan Rough Clustering sangat bergantung pada perhitungan jarak untuk melakukan pengelompokkan data.

Sebelum menjelaskan tentang beberapa macam ukuran jarak, kita mendefinisikan terlebih dahulu yaitu $v_1$ dan $v_2$ menyatakan dua vektor:
$$
v_1 = (x_1, x_2, \ldots, x_n)
$$
$$
v_2 = (y_1, y_2, \ldots, y_n)
$$di mana $x_i$ dan $y_i$ disebut atribut. Ada beberapa ukuran similaritas atau ukuran jarak, di antaranya:

# **1. Minkowski Distance**
Minkowski adalah ukuran jarak umum yang mencakup Euclidean dan Manhattan sebagai kasus khusus.

$$
d(x,y) = \left( \sum_{i=1}^{n} |x_i - y_i|^m \right)^{\frac{1}{m}}, \quad m \ge 1
$$

- $m=1$ → Manhattan  
- $m=2$ → Euclidean  


# **2. Manhattan Distance**
Menghitung jarak sebagai jumlah selisih absolut tiap atribut.

$$
d(x,y) = \sum_{i=1}^{n} |x_i - y_i|
$$

# **3. Euclidean Distance**  
Jarak garis lurus antar dua titik, paling sering digunakan.

$$
d(x,y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

Kelemahan: sensitif terhadap skala besar dan outlier.

# **4. Average Distance**  
Versi rata-rata dari Euclidean untuk mengurangi pengaruh jumlah dimensi.

$$
d(x,y) = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - y_i)^2}
$$

# **5. Weighted Euclidean Distance**  
Memberikan bobot pada setiap atribut.

$$
d(x,y) = \sqrt{\sum_{i=1}^{n} w_i (x_i - y_i)^2}
$$


# **6. Chord Distance**  
Modifikasi Euclidean berbasis normalisasi panjang vektor.

$$
d(x,y) = \sqrt{2 - 2 \frac{\sum x_i y_i}{\|x\| \|y\|}}
$$

# **7. Mahalanobis Distance**  
Memperhitungkan korelasi antar atribut (lebih akurat untuk data kompleks).

$$
d(x,y) = \sqrt{(x - y)^T S^{-1} (x - y)}
$$


# **8. Cosine Similarity**  
Mengukur kemiripan arah (sering dipakai pada teks/dokumen).

$$
\cos(x,y) = \frac{\sum x_i y_i}{\|x\| \|y\|}
$$


# **9. Pearson Correlation**  
Mengukur hubungan linear antar dua data.

$$
r = \frac{\sum (x_i - \mu_x)(y_i - \mu_y)}{\sqrt{\sum (x_i - \mu_x)^2} \sqrt{\sum (y_i - \mu_y)^2}}
$$

Kelemahan: sensitif terhadap outlier.

##**Jarak Atribut Binary**

# **Biner Simetris**
Biner simetris adalah atribut biner di mana kedua nilai (0 dan 1) sama pentingnya. Nilai 1 tidak lebih penting dari 0, begitupun sebaliknya. Rumus menghitung jarak data dengan atribut biner simetris sebagai berikut:



$$
d(x,y) = \sum_{i=1}^{n} |x_i - y_i|
$$

Dimana:
- $x$ merupakan data pertama  
- $y$ merupakan data kedua  
- $x_i$ merupakan fitur ke-$i$ dari data pertama  
- $y_i$ merupakan fitur ke-$i$ dari data kedua  

# **Mengukur Jarak Tipe Categorical (Nominal)**
Digunakan untuk data dengan atribut kategori (misalnya: warna, jenis, dll).

# **1. Overlay Metric (OM)**
Menghitung jarak dengan cara membandingkan setiap atribut.

$$
d(x,y) = \sum_{i=1}^{n} \delta(a_i(x), a_i(y))
$$

- $\delta = 0$ jika sama  
- $\delta = 1$ jika berbeda  

Intinya: hitung berapa banyak atribut yang berbeda.

# **2. Value Difference Metric (VDM)**  
Mengukur jarak berdasarkan perbedaan probabilitas kelas dari tiap nilai atribut.

$$
d(x,y) = \sum_{i=1}^{n} \sum_{c=1}^{C} |P(c|a_i(x)) - P(c|a_i(y))|
$$

Intinya: dua nilai dianggap mirip jika sering muncul pada kelas yang sama.


#**3. Short and Fukunaga Metric (SFM)**  
Mengukur jarak berdasarkan probabilitas keseluruhan data.

$$
d(x,y) = \sum_{c=1}^{C} |P(c|x) - P(c|y)|
$$

Probabilitas biasanya dihitung dengan Naive Bayes.

#**4. Minimum Risk Metric (MRM)**  
Mengukur jarak berdasarkan risiko kesalahan klasifikasi.

$$
d(x,y) = \sum_{c=1}^{C} P(c|x)(1 - P(c|y))
$$

Intinya: semakin kecil risiko salah klasifikasi, semakin mirip datanya.

#**Mengukur Jarak Tipe Ordinal**  

Atribut ordinal adalah data yang memiliki urutan (ranking), misalnya: kecil < sedang < besar.  
Namun, jarak antar nilai tidak diketahui pasti.


Langkah Menghitung Jarak:

#**1. Ubah nilai kategori menjadi peringkat**  
Misalnya:
kecil = 1, sedang = 2, besar = 3  

$$
r_{if} \in \{1, 2, ..., M_f\}
$$


#**2. Normalisasi peringkat ke rentang [0,1]**  
Agar semua atribut memiliki skala yang sama:

$$
z_{if} = \frac{r_{if} - 1}{M_f - 1}
$$


#**3. Hitung jarak seperti data numerik**  
Setelah menjadi $z_{if}$, gunakan jarak seperti Euclidean atau Manhattan.


Intinya:  
- Data ordinal → ubah ke ranking  
- Ranking → dinormalisasi  
- Setelah itu → dihitung seperti data numerik

# **Menghitung Jarak Tipe Campuran**
Digunakan untuk data yang memiliki berbagai tipe atribut:
numerik, nominal, biner, dan ordinal.

#**Rumus Umum:**

$$
d(i,j) = \frac{\sum_{f=1}^{p} \delta_{ij}^{(f)} \, d_{ij}^{(f)}}{\sum_{f=1}^{p} \delta_{ij}^{(f)}}
$$


#**Keterangan:**
- $p$ = jumlah atribut  
- $\delta_{ij}^{(f)} = 0$ jika data hilang / tidak valid  
- $\delta_{ij}^{(f)} = 1$ jika data valid  
- $d_{ij}^{(f)}$ = jarak pada atribut ke-$f$


#**Cara Hitung Tiap Tipe Atribut:**

#1. Numerik  
$$
d_{ij}^{(f)} = \frac{|x_{if} - x_{jf}|}{\max(x_f) - \min(x_f)}
$$
Selisih dinormalisasi (agar skala sama)



#2. Nominal / Biner  
$$
d_{ij}^{(f)} =
\begin{cases}
0, & \text{jika sama} \\
1, & \text{jika berbeda}
\end{cases}
$$

Sama = 0, beda = 1



#3. Ordinal  
Ubah ke ranking lalu normalisasi:

$$
z_{if} = \frac{r_{if} - 1}{M_f - 1}
$$

Setelah itu diperlakukan seperti numerik



Intinya:
- Semua tipe data diubah ke skala yang sama (0–1)
- Hitung jarak tiap atribut
- Ambil rata-rata → jadi jarak total

Ini sering disebut juga sebagai Gower Distance

## Euclidean Distance (Normalized)

Jarak Euclidean adalah panjang garis lurus antara dua titik dalam ruang multidimensi. Rumus umumnya adalah:

$$
d_{Euclidean}(p,q) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2}
$$

Jika diterapkan pada fitur numerik **StudentsPerformance** dengan normalisasi berbasis Range $(R_k)$:

$$
d_{Euclidean\_Norm} =
\sqrt{
\left(\frac{Math_p - Math_q}{R_{Math}}\right)^2 +
\left(\frac{Reading_p - Reading_q}{R_{Reading}}\right)^2 +
\left(\frac{Writing_p - Writing_q}{R_{Writing}}\right)^2
}
$$

---

### Parameter Range $(R_k)$:

- $R_{Math} = 100$
- $R_{Reading} = 100$
- $R_{Writing} = 100$

Range ini diperoleh dari rentang nilai ujian siswa yang berada pada skala **0–100**.

---

### Contoh Data

Misalkan diambil dua siswa:

| Data | Math Score | Reading Score | Writing Score |
|-----|------------|---------------|---------------|
| **P0** | 72 | 72 | 74 |
| **P1** | 69 | 90 | 88 |

---

### Contoh Perhitungan

$$
d =
\sqrt{
\left(\frac{72-69}{100}\right)^2 +
\left(\frac{72-90}{100}\right)^2 +
\left(\frac{74-88}{100}\right)^2
}
$$

$$
d =
\sqrt{
(0.03)^2 + (-0.18)^2 + (-0.14)^2
}
$$

$$
d =
\sqrt{0.0009 + 0.0324 + 0.0196}
$$

$$
d =
\sqrt{0.0529}
$$

$$
d \approx 0.230
$$

```python
import numpy as np
import pandas as pd

cols = ['math score','reading score','writing score']

p1 = pd.Series([72,72,74], index=cols)
p2 = pd.Series([69,90,88], index=cols)

ranges = 100

euclidean = np.sqrt(np.sum(((p1 - p2) / ranges)**2))

print(f"Jarak Euclidean (Normalized) : {euclidean:.4f}")
```

d(p,q) =

## Ordinal

Pada perhitungan jarak untuk tipe data **ordinal**, saya menggunakan data kolom `parental level of education` untuk menghitung jarak antara dua siswa (P0 dan P1) menggunakan ukuran jarak yang telah dijelaskan sebelumnya.

## Perhitungan Jarak Ordinal (Parental Level of Education)

Pada fitur `parental level of education`, terdapat beberapa tingkatan pendidikan orang tua. Kita akan mengikuti prosedur **normalisasi peringkat** agar nilainya berada dalam rentang **[0,1]**.

Urutan tingkat pendidikan dapat ditetapkan sebagai berikut:

1. some high school  
2. high school  
3. some college  
4. associate's degree  
5. bachelor's degree  
6. master's degree  


# **1. Menentukan Peringkat $(r_{if})$ dan Nilai Maksimum $(M_f)$**

Berdasarkan dataset, terdapat **6 tingkatan pendidikan**, sehingga nilai maksimum:

$$
M_f = 6
$$

Misalnya dua siswa:

- **P0:** parental level of education = bachelor's degree  
- **P1:** parental level of education = some college  

Maka peringkatnya:

- $r_{0,edu} = 5$  
- $r_{1,edu} = 3$



# **2. Normalisasi Nilai $(z_{if})$**

Menggunakan rumus:

$$
z_{if} = \frac{r_{if} - 1}{M_f - 1}
$$


# **Perhitungan Normalisasi**

Untuk siswa **P0**:

$$
z_{0} = \frac{5-1}{6-1} = \frac{4}{5} = 0.8
$$

Untuk siswa **P1**:

$$
z_{1} = \frac{3-1}{6-1} = \frac{2}{5} = 0.4
$$



# **3. Menghitung Jarak Manhattan pada Nilai $z_{if}$**

Jarak ordinal antara dua siswa P0 dan P1 dihitung sebagai selisih absolut dari nilai yang telah dinormalisasi:

$$
d_{Ordinal}(p,q) = |z_0 - z_1|
$$

Berdasarkan hasil normalisasi sebelumnya:

$$
z_0 = 0.8
$$

$$
z_1 = 0.4
$$

Sehingga:

$$
d_{Ordinal} = |0.8 - 0.4|
$$

$$
d_{Ordinal} = 0.4
$$

```python
edu_rank = {
    "some high school": 1,
    "high school": 2,
    "some college": 3,
    "associate's degree": 4,
    "bachelor's degree": 5,
    "master's degree": 6
}

p1_class = edu_rank["bachelor's degree"]   # 5
p2_class = edu_rank["some college"]        # 3

M = max(edu_rank.values())

z1 = (p1_class - 1) / (M - 1)
z2 = (p2_class - 1) / (M - 1)

dist_ordinal = abs(z1 - z2)

print(f"Jarak Ordinal (Manhattan) : {dist_ordinal:.4f}")
```

## **Gower’s Similarity Coefficient (Tipe Campuran)**

Berdasarkan rumus Gower, kita akan menjumlahkan seluruh kontribusi jarak dari setiap fitur $(d_{ij}^{(f)})$ lalu membaginya dengan jumlah fitur $(p)$.

### **1. Rekapitulasi Kontribusi Jarak $(d_{ij}^{(f)})$**

| Fitur (f) | Tipe | Nilai P0 | Nilai P1 | Perhitungan | Skor $(d_{ij}^{(f)})$ |
|---|---|---|---|---|---|
| math score | Numerik | 72 | 69 | $\frac{|72-69|}{100}$ | 0.03 |
| reading score | Numerik | 72 | 90 | $\frac{|72-90|}{100}$ | 0.18 |

## **2. Substitusi ke Rumus Gower**

Sesuai rumus:

$$
d(i,j) =
\frac{\sum_{f=1}^{p} \delta_{ij}^{(f)} d_{ij}^{(f)}}
{\sum_{f=1}^{p} \delta_{ij}^{(f)}}
$$

Karena semua data tersedia $(\delta_{ij}=1)$ dan total fitur $(p)$ adalah **6**, maka:

$$
d(0,1) =
\frac{
0.03 + 0.18 + 0.14 + 0 + 1 + 0.4
}{6}
$$

$$
d(0,1) =
\frac{1.75}{6}
$$

$$
d(0,1) \approx 0.2917
$$

```python
# ambil dua data
p1 = df.loc[0]
p2 = df.loc[1]

# --- Numerik ---
dist_math = abs(p1['math score'] - p2['math score']) / 100
dist_reading = abs(p1['reading score'] - p2['reading score']) / 100
dist_writing = abs(p1['writing score'] - p2['writing score']) / 100

# --- Binary ---
dist_gender = 0 if p1['gender'] == p2['gender'] else 1

# --- Nominal ---
dist_lunch = 0 if p1['lunch'] == p2['lunch'] else 1

# --- Ordinal ---
edu_rank = {
    "some high school":1,
    "high school":2,
    "some college":3,
    "associate's degree":4,
    "bachelor's degree":5,
    "master's degree":6
}

r1 = edu_rank[p1['parental level of education']]
r2 = edu_rank[p2['parental level of education']]

M = max(edu_rank.values())

z1 = (r1 - 1) / (M - 1)
z2 = (r2 - 1) / (M - 1)

dist_edu = abs(z1 - z2)
```

```python
import pandas as pd

p1 = pd.Series({
    'math score': 72,
    'reading score': 72,
    'writing score': 74,
    'gender': 'female',
    'lunch': 'standard',
    'parental level of education': "bachelor's degree"
})

p2 = pd.Series({
    'math score': 69,
    'reading score': 90,
    'writing score': 88,
    'gender': 'female',
    'lunch': 'free/reduced',
    'parental level of education': 'some college'
})

dist_math = abs(p1['math score'] - p2['math score']) / 100
dist_reading = abs(p1['reading score'] - p2['reading score']) / 100
dist_writing = abs(p1['writing score'] - p2['writing score']) / 100

dist_gender = 0 if p1['gender'] == p2['gender'] else 1
dist_lunch = 0 if p1['lunch'] == p2['lunch'] else 1

edu_rank = {
    "some high school": 1,
    "high school": 2,
    "some college": 3,
    "associate's degree": 4,
    "bachelor's degree": 5,
    "master's degree": 6
}

r1 = edu_rank[p1['parental level of education']]
r2 = edu_rank[p2['parental level of education']]

M = max(edu_rank.values())
z1 = (r1 - 1) / (M - 1)
z2 = (r2 - 1) / (M - 1)
dist_edu = abs(z1 - z2)

list_jarak = [dist_math, dist_reading, dist_writing, dist_gender, dist_lunch, dist_edu]
hasil_gower = sum(list_jarak) / 6

print(f"Total Jarak Gower (P0 vs P1): {hasil_gower:.4f}")
```
